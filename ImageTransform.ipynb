```python


import os
import cv2
import numpy as np
import csv

def load_images(folder_path, label):
    images = []
    labels = []
    
    for image_name in os.listdir(folder_path):
        image_path = os.path.join(folder_path, image_name)
        
        if not image_path.endswith(('.jpg', '.jpeg', '.png')):
            continue
        
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        resized_image = cv2.resize(image, (28, 28))
        
        normalized_image = resized_image / 255.0
        
        images.append(normalized_image.flatten())  
        labels.append(label)
        
    return images, labels

base_path = 'test'

output_csv_path = 'test.csv'

with open(output_csv_path, mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    
    header = ['label'] + [i for i in range(784)]  #
    csv_writer.writerow(header)
    
    label_counter = 0
    
    for folder_name in os.listdir(base_path):
        class_folder_path = os.path.join(base_path, folder_name)
        
        images, labels = load_images(class_folder_path, label_counter)
        
        for i in range(len(images)):
            csv_writer.writerow([labels[i]] + list(images[i]))
        
        label_counter += 1

base_path = 'train'

output_csv_path = 'train.csv'

with open(output_csv_path, mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    
    header = ['label'] + [i for i in range(784)]  #
    csv_writer.writerow(header)
    
    label_counter = 0
    
    for folder_name in os.listdir(base_path):
        class_folder_path = os.path.join(base_path, folder_name)
        
        images, labels = load_images(class_folder_path, label_counter)
        
        for i in range(len(images)):
            csv_writer.writerow([labels[i]] + list(images[i]))
        
        label_counter += 1
```
